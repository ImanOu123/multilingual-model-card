```python
import pandas as pd
df1 = pd.read_csv("growing_dict/final_terms_google.csv").add_prefix("google_")
df2 = pd.read_csv("growing_dict/final_terms_gpt35.csv").add_prefix("gpt35_")
df3 = pd.read_csv("growing_dict/final_terms_claude3.csv").add_prefix("claude3_")
# df = pd.concat([df1, df2], axis=1)
df = pd.concat([df1, df2, df3], axis=1)
col_lst = ["google_English"] + [f"{j}_{i}" for i in ['Arabic', 'Chinese', 'French', 'Japanese', 'Russian'] for j in ['google', 'gpt35', 'claude3']]
df = df[col_lst]

df = df.apply(lambda x: x.str.split(":").str[-1].str.strip())
df.to_csv("growing_dict/final_terms_agg.csv")
df.head(50)
```

create full dictionary

```python
import pandas as pd
import json
data = pd.read_csv("/home/jiaruil5/multilingual/multilingual-model-card/src/dictionary_collection/growing_dict/final_terms_with_context.csv").to_dict(orient='records')
data_trans = [json.loads(i) for i in open("growing_dict/final_terms_agg.jsonl", 'r').readlines()]
new_data = []
for i, j in zip(data, data_trans):
    j['context'] = eval(i['context'])
    j['English'] = i['processed_term']
    new_data.append(j)
    out_f = open("/home/jiaruil5/multilingual/multilingual-model-card/src/dictionary_collection/growing_dict/mturk.json", 'w')
json.dump(new_data, out_f, indent=2, ensure_ascii=False)
out_f.flush()
```
