```python
import spacy
nlp = spacy.load("en_core_web_trf")
def lemmatize_noun(token):
    if token.pos_ == "NOUN" and token.tag_ == "NNS":
        return token.lemma_
    return token.text

def process_text(text):
    doc = nlp(text)
    results = "".join([lemmatize_noun(token) + token.whitespace_  for token in doc])
    return results

df = pd.read_csv("/home/jiaruil5/multilingual/multilingual-model-card/src/dictionary_collection/growing_dict/final_terms_with_context.csv")
df['original_term'] = df['original_term'].apply(lambda x: process_text(x))
df['original_term']
```

```python
import pandas as pd
import json
import random
import re

def get_numeric_lst(x):
    out_str = ""
    for idx, i in enumerate(x):
        out_str += f"{idx + 1}: {i}<br>"
    return out_str.strip()


def add_context(in_lst, original_term):
    out_lst = []
    for i in in_lst:
        out_lst.append(i.replace(original_term, f"<mark>{original_term}</mark>"))
    return out_lst

def format_context(x, original):
    if len(x) > 2:
        return get_numeric_lst(add_context(random.sample(x, 2), original))
    elif len(x) == 2:
        return get_numeric_lst(add_context([random.choice(x)], original))
    else:
        return "None"
    random.choice(x)

import pandas as pd
import json
data = df.to_dict(orient='records')
data_trans = [json.loads(i) for i in open("growing_dict/final_terms_agg.jsonl", 'r').readlines()]
new_data = []
for i, j in zip(data, data_trans):
    j['context'] = eval(i['context'])
    j['English'] = i['original_term']
    new_data.append(j)
    
df = pd.DataFrame().from_records(new_data)
df['formatted_context'] = df.apply(lambda row: format_context(row['context'], row['English']), axis=1)
df
```

```python
df = df[['English', 'formatted_context', 'Arabic', 'Chinese', 'French', 'Japanese', 'Russian']]
df.rename(columns={"formatted_context": 'context'}, inplace=True)
df
```
